```python
from bs4 import BeautifulSoup as bs
import requests

#setup
url = 'https://apnews.com/hub/astronomy'
page = requests.get(url)
soup = bs(page.content, 'lxml')
counter = 1
headlines = []
links = {} # Sets don't allow for duplicate entries
excluded_links = []

# These links are 'trending' and don't relate to our topic
for link in soup.find_all('a', {'class' : 'Link AnClick-TrendingLink'}):
        excluded_links.append(link.get('href'))
        

# Search for the links we want.
for link in soup.find_all('a', {'class' : 'Link'}):
    links[link.get('href')] = link.text

# Remove Header links    
for link in excluded_links:
    if link in links:
        del links[link]
        
print("\n")

for link in links:
    print(link)
    
```

```python
# Get headlines
for link in links:
    # Setup
    url = link
    page = requests.get(url)
    soup = bs(page.content, 'lxml')
    for headline in soup.find_all('h1', {'class' : 'Page-headline'}):
        headlines.append(headline.text)
    
for h in headlines:
    print(h)
```

```python
counter = 1
# Scrap links for info
content = {}

for link in links:
    # Setup
    url = link
    page = requests.get(url)
    soup = bs(page.content, 'lxml')
    
    # Get rid of unwanted tags
    for tag in soup(['script', 'style', 'img', 'figcaption', 'caption']):
      tag.decompose()


    # Find each paragraph and add it to the content set
    for paragraph in soup.find_all('p'):
        content[counter] = paragraph.text
        counter += 1

for c in content:
    print(content[c])

    #for content in soup.find_all('div', {'class' :"Page-storyBody gtmMainScrollContent"}):
      #  print (content.text.strip())
        
```
